### Fundamentação Teórica

As propriedades da diferenciação no tempo, convolução e modulação da Transformada de Fourier são dadas, respectivamente, por:

$$\frac{d}{dt}x(t)\xrightarrow[\;\;\;\;\;\;\;\;]{\mathfrak{F}}\jmath\omega X(\omega) \tag{1}$$  
$$x(t)\ast y(t)\xrightarrow[\;\;\;\;\;\;\;\;]{\mathfrak{F}}X(\omega).Y(\omega) \tag{2}$$  
$$x(t).y(t)\xrightarrow[\;\;\;\;\;\;\;\;]{\mathfrak{F}}\frac{1}{2\pi}X(\omega)\ast Y(\omega) \tag{3}$$

### Exemplo
Seja o circuito RC série, dado a seguir:

<img src="img\RC-serie.png">

A saída $y(t)$ desse circuito, a tensão sobre o capacitor, pode ser dada por:  

$$y(t)=h(t)\ast x(t) \tag{4}$$

Aplicando a propriedade da convolução, temos:  

$$Y(\omega)=H(\omega).X(\omega) \tag{5}$$

Onde $h(t)$ é a **resposta impulsiva** do sistema no domínio do tempo e $H(\omega)$ é a **resposta em frequência** do sistema no domínio da frequência. A partir da `Eq. 5` podemos obter a resposta em frequência em função das transformadas de Fourier do sinal de entrada e do sinal da saída:

$$H(\omega)=\frac{Y(\omega)}{X(\omega)} \tag{6}$$

Aplicando a Lei de *Kirchoff* para Tensões (**LKT**) ao circuito da Figura 1 temos:

$$x(t)-R.i(t)-y(t)=0 \tag{7}$$

$$i(t)=C.\frac{dy(t)}{dt} \tag{8}$$

Assim:

$$x(t)-R.C.\frac{dy(t)}{dt}-y(t)=0$$
$$R.C.\frac{dy(t)}{dt}+y(t)=x(t)$$

Usando a propriedade da diferenciação no tempo (Eq. 1), teremos:

$$R.C.\jmath\omega Y(\omega)+Y(\omega)=X(\omega)$$

Logo, a **Resposta em Frequência** é dada por:

$$H(\omega)=\frac{Y(\omega)}{X(\omega)}=\frac{1}{1+\jmath\omega R.C}$$

```python
%matplotlib inline
from numpy import arange, abs, angle, pi
from matplotlib.pylab import plot, grid, title, subplot, xlabel, ylabel, figure

w = arange(0,2000*pi,10*pi) 	# base de tempo
R = 1e3; C = 250e-9 		# fc = 4000 rad/s = 636,2 Hz
H = 1. / (1. + 1j*w*R*C) 	# Resp. Freq. do circuito RC série

figure(figsize=(14, 6))
subplot(1,2,1); plot(w,abs(H)); grid('on'); 
title(u'Resp. em Frequência')
xlabel('$\omega$ (rad/s)'); ylabel('Magnitude: $|H(\omega)|$')

subplot(1,2,2); plot(w,angle(H)); grid('on') 
title(u'Resp. em Frequência')
xlabel('$\omega$ (rad/s)'); ylabel('Fase: $\\angle{H(\omega)}\;\;$(rad)');
```
